```python
import numpy as np
from automl_estimators import OptunaEstimator

sequence = [
    ["VarianceThreshold", "SelectPercentile"],      #first step will select between VarianceThreshold and SelectPercentile
    ["PCA"],                                        #second step will always use PCA
    ["LogisticRegression", "KNeighborsClassifier"]  #third step will select between LogisticRegression and KNeighborsClassifier
]

est = OptunaEstimator(sequence, 'roc_auc', n_trials=10, n_jobs=1) #select the best pipeline and tune the hyperparameters using roc_auc as the metric

X = np.random.randn(100, 10)
y = np.random.choice([0, 1], 100)

est.fit(X, y)
est.fitted_pipeline_ 
```
